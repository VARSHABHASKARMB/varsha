```python
x = data.iloc[:,0:1].values
y = data.iloc[:,1].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 40, test_size = 0.2, train_size = 0.5)
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
model = linear_regression.fit(x_train,y_train)
predictions = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print('MSE = :',mean_squared_error(y_test,predictions))
print('R squared value = :',r2_score(y_test,predictions))
```

```python
#Splitting the variables as traing and testing 
```

```python
X = data.iloc[:,0:1].values
Y = data.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.5, test_size = 0.2, random_state = 40)
```

```python
print('Row count of X_train table'+ '-'+str(f"{len(X_train):,}"))
print('Row count of Y_train table'+ '-'+str(f"{len(Y_train):,}"))

print('Row count of X_test table'+ '-'+str(f"{len(X_test):,}"))
print('Row count of Y_test table'+ '-'+str(f"{len(X_test):,}"))
```

```python
#importing statsmodels.api library from statmodel package
```

```python
import statsmodels.api as sm
```

```python
#adding a constant to get an intercept
```

```python
X_train_sm = sm.add_constant(X_train)
```

```python
print('Intercept added to the linear regression model - '+str(X_train_sm ))
```

```python
#Fitting the regression line
```

```python
lr = sm.OLS(Y_train,X_train_sm).fit()
lr.params
lr.summary()
```

```python
#Plot the regression line using the equation 
```

```python
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.scatter(X_train, Y_train)
plt.plot(X_train,-8.066e+07 + 4.024e+04*X_train,'r')
plt.show()
```

```python
#Predict the price
```

```python
X_test_sm = sm.add_constant(X_test)
Y_test_pred = lr.predict(X_test_sm)
Y_test_pred
```

```python
r_squared = r2_score(Y_test, Y_test_pred)
r_squared
```
